```python
import pandas as pd

df = pd.read_csv("fraudData.csv")
```

```python
df.columns
```

```python
features = ['step', 'type', 'amount', 'nameOrig', 'oldbalanceOrg', 'newbalanceOrig',
       'nameDest', 'oldbalanceDest', 'newbalanceDest', 'isFraud',
       'isFlaggedFraud']
df_model = df[features]

df_model.head()
```

```python
# Import label encoder 
from sklearn import preprocessing as pp
  
# label_encoder object knows how to understand word labels. 
label_encoder = pp.LabelEncoder() 

# Encode labels in column 'type'. 
df_model['type']= label_encoder.fit_transform(df_model['type']) 
  
df_model['type'].unique() 
```

```python
X = df_model.drop(columns=['nameOrig','nameDest','isFraud'])
```

```python
X.head()
```

```python
y = df_model['isFraud'].tolist()
y
```

```python
#spliting dataset into training(70%) and test(30%) set
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, 
                                                    test_size=0.3, 
                                                    random_state=42)
```

```python
X_train.head()
```

# Applying the Algorithm

```python
from sklearn.tree import DecisionTreeClassifier as DT
from sklearn.ensemble import AdaBoostClassifier

def adaboost(X_train, X_test, y_train):
    model = AdaBoostClassifier(n_estimators=50, random_state=42)
    model.fit(X_train,y_train)
    y_pred = model.predict(X_test) 
    return y_pred

# AdaBoost
y_pred = adaboost(X_train, X_test, y_train)
```

# Evaluations 

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import confusion_matrix 
print(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.metrics import classification_report 
print(classification_report(y_test, y_pred))
```
